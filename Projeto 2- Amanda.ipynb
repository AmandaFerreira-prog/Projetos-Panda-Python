>6. Quais os temas mais comumente tratados nos emails considerados? (para esse caso, considere usar uma nuvem de palavras como gráfico;

```python
sub=df['subject']
assuntos= " ".join(s for s in sub)
 
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["=?UTF-8?B?", "seu", "em", "você", "de", "ao", "os","foi",'e','UTF','B','C3',
                  '?','o','A7','F0','8J','Q','9F','AA','8D','A7a','A3o'])
 
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(assuntos)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);

```

- Na nuvem de palavras vemos que os assuntos mais frequentes, é o facebook e os pins do pinterest, geralmente as contas que são cadastradas nas redes sociais recebem emails todos os dias, por qualquer atividade da conta, isso explica o porque desse assunto ser tão tratado no email.
