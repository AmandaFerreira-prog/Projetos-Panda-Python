Questão 1:

Diferenças entre a lista, um dicionário, um array numpy e uma serie pandas:

Uma lista (list) em Python é uma sequência ou coleção ordenada de valores. Cada valor na lista é identificado por um índice que só pode ser um número inteiro.

Um	dicionário	ao contrário das lista é uma sequencia não ordenada de valores, e é uma ED em que os elementos são pares chave:valor.	A chave	(key) identifica um	 item e	o valor armazena seu conteúdo. Qualquer valor armazenado pode ser recuperado de	forma extremamente rápida através de sua chave que pode ser de qualquer tipo, inteiro,string,etc.

Já os Arrays Numpy possui similaridades com as listas mas possui alguns poderes a mais, como a capacidade de vetorização, que permite que qualquer operação seja feita sem a necessidade de um laço para acessar seus elementos,tornando a operação mais eficiente.

Já as Series pandas são matrizes unidimensionais rotuladas capazes de armazenar dados de qualquer tipo (inteiro, corda, flutuação, objetos python, etc.). Os rótulos dos eixos são chamados coletivamente de índices. Uma Série suporta tanto a indexação inteira, quanto a baseada em rótulo e fornece uma série de métodos para executar operações envolvendo o índice. Ou seja as series é a uniao de todos as coleções acima, se tornando um dos melhores a se trabalhar.

>Questão 8 - Como alterar a ordem das colunas de um dataframe? Para isso, respondam as três questões: 
- (a) Em df, troque as colunas ’a’ e ’c’. 
- (b) Crie uma função genérica para trocar duas colunas, sem codificar os nomes das colunas.
- (c) Classifique as colunas em ordem alfabética inversa, ou seja, colume ’e’ primeiro até a coluna ’a’ last. 2 df = pd.DataFrame(np.arange(20).reshape(−1, 5), columns=list('abcde'))

```python
#Questão 8- Letra (A):
df = pd.DataFrame(np.arange(20).reshape(-1,5), columns=list('abcde'))
df.reindex(columns=['c','b','a','d','e'])
```

```python
#Questão 8 - letra (B)

def alternarcolunas(df,col1,col2):
    l=list(df.columns)
    col1,col2=l.index(col1),l.index(col2)
    l[col1],l[col2]=l[col2],l[col1]
    df=df[l]
    return df
df = pd.DataFrame(np.arange(20).reshape(-1,5), columns=list('abcde'))
alternarcolunas(df,'a','c')


    
```

```python
#Questão 8 - Letra (C)

df.reindex(columns=['e','d','c','b','a'])
```

>Questão 11 - Obtenha as duas últimas linhas de df cuja soma da linha é maior que 100. df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(−1, 4))

```python
#Questão 11
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1,4))
soma=df.sum(axis=1)
cont=0
for n in range(14,-1,-1):
    if soma.loc[n]>100:
        print('linha:', n,' soma:', soma.loc[n])
        cont+=1
    if cont==2:
        break
        
```
